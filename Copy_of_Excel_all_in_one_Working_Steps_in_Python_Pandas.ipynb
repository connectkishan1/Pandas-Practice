# Step3: Data Preprocessing

````
df_drop_missing=df.drop(df.columns[df.isnull().mean() > 0.35],axis=1)
df_drop_missing
````

````
df_replace=df_drop_missing.fillna(df_drop_missing['Site I&C'].mean())
df_replace
````

````
df_replace["Name"].apply(lambda x:x.upper())
````

````
df.Date=pd.to_datetime(df.Date)
df.Date
````

````
df_replace[['year','month','day']]=df_replace.Date.apply(lambda x: pd.Series(x.strftime("%y,%m,%d").split(",")))
df1=df_replace[['year','month','day']]
df1
````

````
df_replace[['year']]=df_replace.Date.apply(lambda x: pd.Series(x.strftime("%Y").split(",")))
df2=df_replace[['year']]
df2
````

````
df_replace[['year','month','day']]=df1
df_replace
````

````
df_replacing_rows_value1=df_replace[df_replace['Site Audits']<df_replace['Site I&C']]
df_replacing_rows_value1
````

````
df_replacing_rows_value2=df_replace[df_replace['Site Audits']>df_replace['Site I&C']]
df_replacing_rows_value2["Site Audits"]=df_replacing_rows_value1["Site Audits"].mean()
df_replacing_rows_value2
````

````
df_replacing_rows_value=pd.concat([df_replacing_rows_value1,df_replacing_rows_value2])
df_replacing_rows_value
````

````
df_replacing_rows_value["audit%"]=df_replacing_rows_value['Site Audits']/df_replacing_rows_value['Site I&C']*100
df_replacing_rows_value#df_with_added_column using airthmaic operation
````

````
#df_replacing_rows_value.insert(8,'audit%',df_replacing_rows_value['audit%'],allow_duplicates=True) 
````

````
order = [0,1,2,3,4,5,6,11,7,8,9,10]
df_final = df_replacing_rows_value[[df_replacing_rows_value.columns[i] for i in order]]
df_final
````

````
df55 = df_replacing_rows_value.iloc[:, [0,1,2,3,4,5,6,11,7,8,9]]
df55
````
